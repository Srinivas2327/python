# input:1234
# output:6 (2+4)

```
def addevendigit(n):
    sum=0
    while n!=0:
      r=n%10
      if r %2==0:
          sum=sum+r
      n=n//10
    return sum
addevendigit(1234)
```

```
def largest(n):
  l=0
  while n!=0:
    r=n%10
    if r>l:
       l=r
    n=n//10
  return l
largest(19528)
    
```

# input: 121-----121--yes
#   123----321---no

```
def palin(n):
    rev=0
    temp=n
    while n!=0:
      r=n%10
      rev =rev*10+r
      n=n//10
    if temp==rev:
       return "yes"
    else:
       return "no"
print(palin(113))
print(palin(121))
```

```
def natural(n):
    for i in range(1,n+1):
       print(i,end=" ")
natural(10)
```

```
def natural(s,r):
    for i in range(s,r+1):
       print(i,end=" ")
natural(1,25)
```

```
def natural(s,r):
    for i in range(s,r+1,4):
       print(i,end=" ")
natural(1,25)
```

```
def natural(s,r):
    for i in range(s,r+1,3):
       print(i,end=" ")
natural(1,10)
```

```
def natural(s,e):
    for i in range(e,s-1,-1):
       print(i,end=" ")
natural(1,25)
```
